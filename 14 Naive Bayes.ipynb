### Prepare a classification model using Naive Bayes for salary data 

```python
import pandas as pd
from sklearn import preprocessing
from sklearn.naive_bayes import CategoricalNB
from sklearn import preprocessing
import numpy as np
```

```python
data_test = pd.read_csv('SalaryData_Test.csv')
data_test
```

```python
data_train=pd.read_csv('SalaryData_Train.csv')
data_train
```

```python
total_data = [data_test, data_train]

sal_data = pd.concat(total_data)
sal_data
```

```python
sal_data.head()
```

```python
def swap_columns(sal_data, Salary, age):
    col_list = list(sal_data.columns)
    x, y = col_list.index(Salary), col_list.index(age)
    col_list[y], col_list[x] = col_list[x], col_list[y]
    sal_data = sal_data[col_list]
    return sal_data
sal_data1 = swap_columns(sal_data, 'Salary', 'age')
print (sal_data1)
```

```python
sal_data1
```

```python
sal_data2 = sal_data1.drop('native',axis=1 )
sal_data2
```

```python

encode = preprocessing.LabelEncoder()
wc_encoded = encode.fit_transform(sal_data2['workclass'])
label = encode.fit_transform(sal_data2['Salary'])
edu_encoded = encode.fit_transform(sal_data2['education'])
edu1_encoded = encode.fit_transform(sal_data2['educationno'])
mar_encoded = encode.fit_transform(sal_data2['maritalstatus'])
occu_encoded = encode.fit_transform(sal_data2['occupation'])
rel_encoded =encode.fit_transform(sal_data2['relationship'])
race_encoded =encode.fit_transform(sal_data2['race'])
sex_encoded = encode.fit_transform(sal_data2['sex'])
cg_encoded = encode.fit_transform(sal_data2['capitalgain'])
cl_encoded = encode.fit_transform(sal_data2['capitalloss'])
hpw_encoded = encode.fit_transform(sal_data2['hoursperweek'])
age_encoded = encode.fit_transform(sal_data2['age'])
```

```python

print ('workclass:', wc_encoded)
print ('education:', edu_encoded)
print ('educationno:', edu1_encoded)
print ('maritalstatus:', mar_encoded)
print ('occupation:', occu_encoded)
print ('relationship:', rel_encoded)
print ('race:', race_encoded)
print ('sex:', sex_encoded)
print ('capitalgain:', cg_encoded)
print ('capitalloss:', cl_encoded)
print ('age:', age_encoded)
print ('hoursperweek:', hpw_encoded)

print('Salary:', label)
```

```python
X=list(zip(wc_encoded, edu_encoded, edu1_encoded, mar_encoded,occu_encoded,rel_encoded,race_encoded,sex_encoded,cg_encoded,cl_encoded,age_encoded, hpw_encoded))
y = list(label)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB as CB
from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import GaussianNB as GB
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
X_train, y_train, X_test, y_test
```

```python
model = GB()

model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
print('predicted value:',predicted)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predicted))
```

### we can see the predicted score as 81% and hence can conclude it to be a good model
